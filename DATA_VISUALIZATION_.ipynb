```python
import matplotlib.pyplot as plt
months=['Jan','Feb','Mar','Apr','May','Jun']
energy_consumption=[1200,1300,1100,1500,1400,1600]
plt.plot(months,energy_consumption,marker='o',color='b',linestyle='--')
plt.xlabel('Months')
plt.ylabel('Energy Consumption')
plt.title('Energy Consumption Over Time')
plt.show()
```

```python
energy_sources=['Solar','Wind','Hydro','Biomass']
energy_values=[1200,1300,1100,1500]
plt.bar(energy_sources,energy_values,color=['b','g','r','y'])
plt.xlabel('Energy Sources')
plt.ylabel('Energy Consumption')
plt.title('Energy Consumption by Source')
```

```python
plt.pie(energy_values,labels=energy_sources,colors=['b','g','r','y'],autopct='%1.1f%%')
plt.axis('equal')
plt.title('Energy Consumption by Source')
plt.show()
```

```python
plt.scatter(months,energy_consumption,color='r')
plt.xlabel('Months')
plt.ylabel('Energy Consumption')
plt.title('Energy Consumption Over Time')
plt.show()
```

```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

data = {
    "Solar": [1200,1500,1300],
    "Wind": [3400,3600,3200],
    "Hydropower": [2900,3100,2800],
    "Biomass": [2500,2700,2400]

}
df=pd.DataFrame(data)
sns.pairplot(data=df)
plt.show()
```

```python
correlation_matrix = df.corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm',linewidths=0.5)
plt.title('Correalation Between energy sources')
plt.show()
```

```python
sns.boxplot(data=df)
plt.title('Distribution of energy consumption bty sources')
plt.show()
```

```python
sns.violinplot(data=df)
plt.title('Violen plot Distribution of energy consumption bty sources')
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

energy_values =[100,200,300,400,500]
carbon_emissions=[10,20,25,40,50]
df_reg=pd.DataFrame({
    'Energy Consumption':energy_values,
    'Carbon Emissions':carbon_emissions
})
sns.regplot(x='Energy Consumption',y='Carbon Emissions',data=df_reg)
plt.title('Relationship between Energy Consumption and Carbon Emissions')
```

```python
from mpl_toolkits import mplot3d
import numpy as np
import matplotlib.pyplot as plt

fig=plt.figure()
ax=plt.axes(projection='3d')

z=np.linspace(0,1,1000)
x=z*np.sin(25*z)
y=z*np.cos(25*z)

ax.plot3D(x,y,z,'green')
ax.set_title('3D Line Plot')
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Import the IRIS dataset
iris = pd.read_csv('IRIS.csv')

# Create a pair plot
sns.pairplot(iris, hue='species')
plt.show()
```

```python
# Create a box plot
sns.boxplot(x='species', y='sepal_length', data=iris)
plt.show()
```

```python
# Create a violin plot
sns.violinplot(x='species', y='sepal_length', data=iris)
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt



# Select numerical columns
numerical_iris = iris.select_dtypes(include=['number'])

# Calculate correlation matrix for numerical columns
correlation_matrix = numerical_iris.corr()

# Create heatmap
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.show()
```

```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import LabelEncoder

iris = sns.load_dataset('iris')

iris_encoded = iris.copy()

le = LabelEncoder()

iris_encoded['species'] = le.fit_transform(iris_encoded['species'])

print(iris_encoded.head())

correlation_matrix = iris_encoded.corr()

sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Heatmap of Iris Features and Species')
plt.show()
```
